```python
df_2 = pd.read_csv (r'C:\Users\20191169\Desktop\Courses\Year 3\Q1\Data Challenge 3\all_weirs_clean.csv')
df_2.head()
```

```python
sns.distplot([float(i) for i in df_2["VERSCHIL"]])
plt.title("Distribution of Verschil after outlier removal")
plt.xlabel("VERSCHIL")
plt.ylabel("Probability Density")
plt.show()
```

```python
sns.distplot([float(i) for i in df_2["Q"]])
plt.title("Distribution of Q after outlier removal")
plt.xlabel("Q")
plt.ylabel("Probability density")
plt.show()
```

```python
import seaborn as sns

clean_data1 = df_2.set_index('TIME')

# stripplot for verschil
plt.figure(figsize=(25,10))

ax = sns.stripplot(data = clean_data1, x='STUWVAK', y='VERSCHIL', jitter = 0.2)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
ax.set_title('Stripplot of Verschil for Each Stuwvak', size = 40, weight = 'bold');
plt.show()
```

```python
plt.figure(figsize=(25,10))

ax = sns.stripplot(data = clean_data1, x='STUWVAK', y='Q', jitter = 0.2)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
ax.set_title('Stripplot of Q for Each Stuwvak', size = 40, weight = 'bold');
plt.show()
```
